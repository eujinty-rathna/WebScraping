```python
def get_dic(url):
    dic ={}
    split_url = re.split('//?',url)
    scheme = split_url[0].rstrip(':')
    domain = split_url[1].strip()
    dic['url'] = url
    dic['scheme'] = scheme
    dic['domain'] = domain
    dic['brand_name'] = re.search('(\w+)-',split_url[2]).group(1)
    dic['sku_number'] = re.search('(\d+-\d+)-',split_url[2]).group(1)
    dic['color'] = re.search('\d-([A-z]+)-',split_url[2]).group(1)
    dic['occasion'] = re.search(fr"{dic['color']}-(\w+)",split_url[2]).group(1)
    dic['type']=re.search(fr"{dic['occasion']}-(.*).html",split_url[2]).group(1)
    dic['extenstion'] = re.search('\.(\w+)',split_url[2]).group(1)
    return dic
```
